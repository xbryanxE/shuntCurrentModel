```python
import numpy as np
import matplotlib.pyplot as plt
```

## 1. Tafel kinetics

```python
from TafelKinetics import Tafel
```

### 1.1. Overpotentials as input

```python
# Define the parameters for the Tafel equation
b = 0.06 # V/decade
j0 = 1.2e-4 # A/cm^2
# Initialize the Butler-Volmer instance
bv = Tafel(j0, b)
# Define the overpotential range
eta = np.linspace(-0.25, 0.4, 100) # V
# Calculate the current density using the Tafel equation
j = bv.current_density(eta)
# Plot the results
plt.figure(figsize=(8, 6))
plt.plot(eta, j, label='Tafel Current Density')
plt.xlabel('Overpotential (V)')
plt.ylabel('Current Density (A/cm^2)')
plt.title('Tafel Equation')
plt.axhline(0, color='black', lw=0.5, ls='--')
plt.axvline(0, color='black', lw=0.5, ls='--')
plt.grid()
```

### 1.2. Curent density as input

```python
j = np.linspace(-1e-4, 5e-3, 100) # A/cm^2
# Calculate the overpotential using the Tafel equation
eta = bv.overpotential(j)
# Plot the results
plt.figure(figsize=(8, 6))
plt.plot(j, eta, label='Tafel Overpotential')
plt.xlabel('Current Density (A/cm^2)')
plt.ylabel('Overpotential (V)')
plt.title('Tafel Overpotential')
plt.axhline(0, color='black', lw=0.5, ls='--')
plt.axvline(0, color='black', lw=0.5, ls='--')
plt.grid()
```

## 2. Butler-Volmer Kinetics

```python
from ButlerVolmer import ButlerVolmer
```

### 2.1. Overpotentials as inputs

```python
# Define the parameters for the Butler-Volmer equation
ba = 0.04 # V/decade
bc = 0.12 # V/decade
j0 = 1.2e-4 # A/cm^2
# Initialize the Butler-Volmer instance
bv = ButlerVolmer(j0, ba, bc)
# Define the overpotential range
eta = np.linspace(-0.25, 0.08, 100) # V
# Calculate the current density using the Butler-Volmer equation
j = bv.current_density(eta)
# Plot the results
plt.figure(figsize=(8, 6))
plt.plot(eta, j, label='Butler-Volmer Current Density')
plt.xlabel('Overpotential (V)')
plt.ylabel('Current Density (A/cm^2)')
plt.title('Butler-Volmer Equation')
plt.axhline(0, color='black', lw=0.5, ls='--')
plt.axvline(0, color='black', lw=0.5, ls='--')
plt.grid()
```

### 2.2. current density as input

```python
j = np.linspace(-5e-4, 5e-4, 100) # A/cm^2
# Calculate the overpotential using the Butler-Volmer equation
eta = bv.overpotential(j)
# Plot the results
plt.figure(figsize=(8, 6))
plt.plot(j, eta, label='Butler-Volmer Overpotential')
plt.xlabel('Current Density (A/cm^2)')
plt.ylabel('Overpotential (V)')
plt.title('Butler-Volmer Overpotential')
plt.axhline(0, color='black', lw=0.5, ls='--')
plt.axvline(0, color='black', lw=0.5, ls='--')
plt.grid()
```
